# EDA(Exploratory Data Analysis)

- EDA : 탐색적 자료 분석, 즉 수집한 자료를 `다양한 각도에서 관찰하고 이해`하는 과정으로 그래프나 통계적 방법을 이용해서 자료를 직관적으로 파악하는 과정을 말한다. `자료의 분포와 통계를 파악하여 자료가 갖고 있는 특성을 이해`하고, 잠재적인 문제를 발견해서 분석전에 발견이 어려운 다양한 문제점을 발견하고 이를 바탕으로 기존의 가설을 수정하거나 새로운 방향의 가설을 세울 수 있다. 
- EDA는 분석에 필요한 자료를 이해하고 자료의 전처리 그리고 정제된 데이터셋을 대상으로 변수 간의 특성을 분석하는 데 필요하다. 
  - 1) 분석의 목적과 변수의 특징 확인
    - 수집된 자료를 대상으로 셀 수 없는 변수와 셀 수 있는 변수를 구분한다.
  - 2) 자료 확인 및 전처리
    - 수집된 자료를 확인하고 결측치와 이상치를 확인하고 정제한다.
  - 3) 자료의 각 변수 관찰
    - 대표값이나 산포도를 이용한 통계 조사 및 이산변수와 연속변수 시각화
  - 4) 변수 간의 관계에 초점을 맞춰 패턴 발견
    - 변수와 변수 간의 상관관계이나 고급 시각화 도구를 이용한 변수 간의 패턴 발견

- 결측치 처리
  - na.rm=T 또는 na.omit() 함수를 통해 결측치를 제거한 후 유효한 자료만을 대상으로 연산할 수도 있다. 하지만 무조건 결측치를 제거한다면 그로 인하여 중요한 데이터가 손실될 수 있다. 따라서 결측치를 처리하는 방법에는 결측치를 제거하는 방법도 있지만 결측치를 제거하지 않고 다른 값으로 대체하는 방법도 고려해볼 수 있다. 

- 결측치 확인

```r
summary(dataset$price)
```

- 결측치가 30개 있다.

```r
sum(dataset$price) # error
sum(dataset$price,na.rm=T) # 무시
sum(na.omit(dataset$price))
# na.omit() 함수는 결측치를 제거한 후 남은 것을 반환해준다.
```

- 결측치 대체 (0으로 대체하거나 평균으로 대체한다.)

```r
# NA이면 0으로 대체
ifelse(!is.na(dataset$price),dataset$price,0)
```

```r
# 결측치를 평균으로 대체
ifelse(!is.na(dataset$price),dataset$price,round(mean(dataset$price,na.rm=T),2))
```

```r
# 이렇게 기존 df에 새로운 칼럼을 만들어줄 수 있음
dataset$newprice = ifelse(!is.na(dataset$price),dataset$price,round(mean(dataset$price,na.rm=T),2))
```

- 극단치 처리
  - 극단치 : 표본 중 다른 대상들과 확연히 구분되는 통계적 관측치를 의미한다. 즉, 변수의 분포에서 비정상적으로 분포를 벗어난 값을 극단치(outlier)라고 한다. 예를 들면 나이의 분포는 0세 이상 100세 사이의 분포를 보이는 것이 일반적이지만 -2 또는 230과 같은 비정상적인 수치가 보인다면 이는 극단치에 해당한다. 또한 변수와의 상관관계에 있어서 높은 직급의 경력사원이 직급이나 낮은 일반 사원보다 급여가 월등하게 적다면 이 또한 비정상적인 수치로 의심해볼 수 있다. 

- 범주형 변수 극단치 처리

- 코딩 변경
  - 최초 코딩 내용을 용도에 맞게 변경하는 작업을 의미.데이터 가독성과 척도 변경 그리고 역 코딩 등의 목적으로 수행.

- 1) 가독성을 위한 코딩 변경
  - 일반적으로 데이터는 디지털화하기 위해서 숫자로 코딩한다. 
    - 예를 들면 거주지 관련 칼럼에서 서울은 1, 인천은 2로 코딩한다. 이런 코딩 결과를 대상으로 기술통계분석을 수행하면 1과 2의 숫자를 실제 거주지명으로 표현해야 한다. 실습해보자

```r
head(dataset2)
```

- resident 칼럼을 보면 실제 거주지명이 아닌 1,2,3,4,5로 표현되어 있음을 알 수 있다. 
- dataset2의 resident 칼럼으로 조건식을 지정하여 거주지역별로 실제 거주지명을 resident2 칼럼에 추가한다.

```r
dataset2$resident2[dataset2$resident == 1] = '1.서울특별시'
dataset2$resident2[dataset2$resident == 2] = '2.인천광역시'
dataset2$resident2[dataset2$resident == 3] = '3.대전광역시'
dataset2$resident2[dataset2$resident == 4] = '4.대구광역시'
dataset2$resident2[dataset2$resident == 5] = '5.시구군'
```
