# 데이터 전처리

- 칼럼으로 데이터 정렬

````r
# 데이터 로딩
library(hflights)
library(dplyr)
# 데이터를 년, 월, 출발시간, 도착시간 기준으로 오름차순 정렬
arrange(hflights,Year,Month,DepTime,ArrTime)
# 정렬의 우선순위는 가장 왼쪽의 칼럼부터 시작하여 오른쪽 칼럼 순으로 정렬된다.
````

- 칼럼으로 데이터 검색
  - 대용량의 데이터 셋의 특정 칼럼을 기준으로 데이터를 검색할 수 있다. 
  - 관계형 데이터 베이스에서 데이터 검색을 위한 select 문과 비슷한 기능을 수행하는 select() 함수에 대해서 알아본다.

````r
# 이렇게 4개의 칼럼만 검색한다.
select(hflights,Year,Month,DepTime,ArrTime)
````

````r
# 칼럼의 범위로 검색하기
select(hflights,Year:ArrTime)
# 만약 특정 칼럼 또는 칼럼의 범위를 검색에서 제외하려는 경우 제외하려는 칼럼 이름 또는 범위 앞에 -속성을 지정한다.
````

- 데이터 셋에 칼럼 추가
  - 대용량의 데이터 셋을 대상으로 특정 칼럼을 추가하는 함수에 대해서 알아보자

````r
mutate(hflights,gain=ArrDelay-DepDelay)
# gain이라는 칼럼명을 추가해줄 것이고 수식은 ArrDelay-DepDelay와 같다.
````

````r
# 혹은 다음과 같이 이용할 수도 있겠다.
hflights%>%mutate(gain=ArrDelay-DayOfWeek)
````

- 요약 통계 구하기
  - 대용량의 데이터 셋을 대상으로 특정 칼럼을 대상으로 기술 통계량을 계산하는 관련 함수에 대해서 알아보자

````r
summarise(hflights,avgAIRTIME = mean(AirTime,na.rm=T))
````

````r
hflights%>%summarise(avg=mean(AirTime,na.rm=T))
````

````r
str(hflights)
````

````r
summarise(hflights,count=n(),delay=mean(AirTime,na.rm=T))
# n()은 데이터 셋의 관측치 길이를 구해준다. dplyr 패키지에서 제공해주는 함수이다.
````

- 데이터프레임 병합
  - 서로 다른 데이터프레임을 대상으로 공통 칼럼을 이용하여 하나의 데이터프레임으로 병합하는 함수에 대해서 알아보자

- 공통 변수를 이용하여 내부조인 하기

````r
df1
df2
````

````r
df1=data.frame(x=1:5,y=rnorm(5))
df2=data.frame(x=2:6,y=rnorm(5))
inner_join(df1,df2,by='x')
# 공통으로 사용된 x는 병합을 위해 사용된다.
# 병합의 결과는 양쪽 모두 x가 존재하는 2,3,4,5 관측치를 대상으로 병합된다. 
# inner_join은 내부조인이라고 부른다. 
````

- 공통변수를 이용하여 왼쪽 조인하기

````r
left_join(df1,df2,by='x')
# 왼쪽에 위치한 df1의 x칼럼을 기준으로 오른쪽 df2와 병합한다.
# 만약 오른쪽의 df2에서 해당하는 x의 값이 없는 경우에는 결측치 NA로 둔다. 
# 왿쪽 데이터프레임을 기준으로 병합하기 때문에 왼쪽 조인이라고 부른다.
````

- 공통변수를 이용하여 오른쪽 조인하기

````r
right_join(df1,df2,by='x')
````

- 공통변수를 이용하여 전체 조인하기

````r
full_join(df1,df2,by='x')
````

- 위와는 다른 상황인, 서로 다른 데이터프레임을 대상으로 행 단위 또는 열 단위로 합치는 함수에 대해서 알아보자

````r
bind_cols(df1,df2)
````

- reshape2 패키지 활용
  - reshape2 패키지에서 제공하는 dcast() 함수를 이용하여 긴 형식의 데이터를 넓은 형식의 데이터로 변경한다.

- 긴 형식을 넓은 형식으로 변경

````r
install.packages('reshape2')
````

````r
library(reshape2)
````

```r
wide = dcast(데이터셋, 앞변수~뒷변수, 적용함수)
```
- 이때 dcast() 함수를 실행하면 데이터 셋을 대상으로 앞 변수와 뒷 변수의 값이 같은 경우 측정변수를 대상으로 적용함수가 적용된다. 이때 앞 변수는 각 행을 구성하며, 뒷 변수를 각 열을 구성한다. 실행 후 출력되는 결과 메시지는 어떤 변수가 측정변수로 사용되어 정리 대상 변수가 되었다고 알려주는 메시지이다. 

- 넓은 형식을 긴 형식으로 변경
  - reshape2 패키지에서 제공하는 melt() 함수를 이용하여 넓은 형식의 데이터를 긴 형식의 데이터로 변경할 수 있다. 

```r
melted = melt(데이터 셋, id='칼럼명')
```

- 넓은 형식의 데이터 셋을 대상으로 id='칼럼명'로 지정된 칼럼명을 기준으로 데이터의 구조를 긴 형식으로 변경하는 과정이다.
- 칼럼명을 변경하려면
```r
colnames(melted) = c('a','v',...)
```

````r
data(smiths)
smiths
````

- 현재는 넓은 형식의 데이터 셋이다.

````r
long = melt(smiths, id=1:2)
long
````

- smiths 데이터 셋을 대상으로 첫 번째 칼럼과 두 번째 칼럼을 기준으로 긴 형식으로 구조가 변경되어 4개의 칼럼과 6개의 관측치를 갖는 데이터프레임이 생성된다. 

- 긴 형식을 넓은 형식으로 변경하기

````r
dcast(long,subject+time~...)
````

- 긴 형식으로 변경된 데이터 셋을 대상으로 subject와 time 칼럼을 기준으로 다시 넓은 형식으로 구조를 변경하였다. 

- 3차원 배열 형식으로 변경

````r
data('airquality')
````

````r
str(airquality)
````

- 칼럼 제목을 대문자로 일괄 변경

- melt() 함수를 이용하여 넓은 형식을 긴 형식으로 변경하기

````r
air_melt = melt(airquality, id = c('MONTH','DAY'),na.rm=T)
head(air_melt)
````

- acast() 함수를 이용하여 3차원 구조로 변경하기

````r
names(air_melt) = tolower(names(air_melt))
````

````r
head(air_melt)
````

````r
acast = acast(air_melt,day~month~variable)
acast
````
