# 군집 분석

- 신입사원의 면접시험 결과를 군집 분석

```r
interview = read.csv('interview.csv', header = T)
head(interview,3)
names(interview)
```

```r
# 유클리디안 거리 계산
interview_df = interview[c(2:7)]
idist = dist(interview_df) # 유클리디안 거리 계산
head(idist,3)
```

- 군집 분석에 필요없는 no, 종합점수, 합격여부 칼럼은 생략하고 유클리디안 거리를 계산한다.

```r
# 계층적 군집 분석
hc = hclust(idist)
hc # 계층적 군집 분석 결과 보기
```

```r
# 군집 분석 시각화
plot(hc, hang = -1) # 음수 값 제외
# hang = -1 속성을 추가해주면 덴드로그램에서 음수값을 제외할 수 있다. 

# 군집 단위 테두리를 생성해보자
# 빨간 네모 선 추가해주기
rect.hclust(hc, k=3, border='red') 
```

- 3개의 그룹으로 군집이 형성된 것을 확인할 수 있다. 

```r
# 군집별 특징 보기

# 군집별 서브 셋 만들기
g1 = subset(interview, no == 108 | no == 110 | no == 107 | no == 112 | no == 115)
g2 = subset(interview, no == 102 | no == 101 | no == 104 | no == 106 | no == 113)
g3 = subset(interview, no == 105 | no == 114 | no == 109 | no == 103| no == 111)

# 각 subset의 요약통계량
summary(g1)
summary(g2)
summary(g3)
```

- 이처럼 군집 분석으로 그룹화된 군집은 다변량적 특징이 그룹 내적으로는 동일하고, 외적으로는 이질적인 특성을 갖는다.

- 군집 수 자르기
  - 계층별 군집 분석 결과에서 분석자가 원하는 군집 수만큼 잘라서 인위적으로 군집을 만들 수 있다. 

```r
# 유클리디안 거리 계산
idist = dist(iris[1:4])
# 계층형 군집 분석(클러스트링)
hc = hclust(idist)
plot(hc, hang = -1)
rect.hclust(hc, k = 4, border = 'red')
# 군집 수 = 4개
```

- 군집 수 잘라보자

```r
# 150개의 관측치를 대상으로 3개의 군집 수를 지정하여 ghc 변수에 저장한다.
ghc = cutree(hc, k = 3)
ghc
```

```r
# iris 데이터 셋에 ghc 칼럼 추가
iris$ghc = ghc
# ghc 빈도수
table(iris$ghc)

head(iris, 3)
```

```r
# 요약통계량 구하기
g1 = subset(iris, ghc == 1) # 제 1군집만 subset 작성
summary(g1)
```

- 비계층적 군집 분석
  - 군집의 수가 정해진 상태에서 군집의 중심에서 가장 가까운 개체를 하나씩 포함해 나가는 방법.
  - K-means Clustring이 있다. 
  - 군집의 수를 미리 알고 있는 경우 군집 대상의 분포에 따라 군집의 초기값을 설정해주면 초기값에서 가장 가까운 거리에 있는 대상을 하나씩 더해 가는 방식으로 군집화를 수행하게 된다.
  - 따라서 계층적 군집 분석을 통해 대략적인 군집의 수를 파악하고 이를 초기 군집의 수로 설정하여 비계층적 군집 분석을 수행하는 것이 효과적이다. 

```r
# 단계 1 : 군집 분석에 사용할 변수 추출
library(ggplot2)
data(diamonds)

# 데이터 샘플링
t = sample(1:nrow(diamonds), 1000)
# 표본으로 검정데이터 생성
test = diamonds[t,]
# 군집을 위해서 필요한 변수 추출
mydia = test[c('price','carat','depth','table')]
head(mydia,3)
```

```r
# 단계 2 : 계층적 군집 분석(탐색적 분석)

result = hclust(dist(mydia), method = 'average') # 평균 거리 인용
result

plot(result, hang = -1)
```

```r
# 단계 3 : 비계층적 군집 분석
result2 = kmeans(mydia, 3) # 3개 군집 수 적용
names(result2)
```

```r
result2$centers
```

```r
result2$cluster
```

```r
# 원형 데이터에 군집 수 추가
mydia$cluster = result2$cluster
head(mydia,4)
# 끝에 cluster 칼럼이 추가된 것을 알 수 있음
```

```r
# 단계 4 : 변수 간의 상관계수 보기
cor(mydia[,-5], method = 'pearson')
plot(mydia[,-5])
```

- price 변수에 가장 큰 영향을 미치는 변수는 carat이며, depth는 부(-)의 영향을 미친다. 

```r
# 상관계수를 색상으로 시각화
library(mclust)
library(corrgram)

corrgram(mydia[,-5], upper.panel = panel.conf)
corrgram(mydia[,-5], lower.panel = panel.conf)
```

- 첫 번째 그래프는 상관계수가 위쪽으로 나타나고 두 번째 그래프는 상관계수가 아래쪽으로 나타난다. 

```r
# 비계층적 군집 시각화
plot(mydia$carat, mydia$price, col = mydia$cluster)
# 중심점 표시 추가
points(result2$centers[,c('carat','price')], col = c(3,1,2), pch = 8, cex = 8)
```

- 다이아몬드의 크기가 클수록 대체로 가격이 상승한다는 특징이 있다.
