# 분석 절차와 통계 지식

- 추론통계를 바탕으로 하는 분석 절차
  - 1) 가설검정
    - 실증적인 증명에 앞서 세우는 잠정적인 진술, 통계분석을 통해서 채택 또는 기각될 수 있다.
    - 귀무가설(영가설) - '~관계가, 차이가 없다'등의 부정적 형태로 진술하는 가설($H_0$) 
    - 대립가설(연구가설) - '차이가 있다'등의 형태로 진술하는 가설($H_1$), 검증하고자 하는 현상에 관한 예측
    - 논문에서는 대립가설을 제시하고, 이 가설이 채택 또는 기각되는지는 귀무가설을 통해서 검정한다. 
    - 귀무가설을 입증하기가 훨씬 쉽기 때문에 자신이 내세운 귀무가설이 틀렸음을 통계적인 분석과정을 통해서 입증함으로써 귀무가설과 대립 관계에 있는 연구가설을 채택하고 궁극적으로 연구가설이 사실임을 주장하고자 한다. 
  - 2) 유의수준 결정
    - 유의수준에 기준하여 가설의 채택여부가 결정된다. 유의수준은 $\alpha = 0.5  또는 P < 0.5 $로 표시한다.
    - $\alpha > P$ : 연구가설 채택, 귀무가설 기각
    - 분석 결과가 유의 확률(p)이 유의 수준($\alpha$)보다 작은 경우 귀무가설을 기각, 그렇지 않으면 기각하지 못한다. 
    - 즉 유의확률이 유의수준보다 작다는 것은 관측치가 귀무가설의 기각역에 있다는 것을 의미한다. 
  - 3) 측정도구 선정
    - 가설에 나오는 변수를 무엇으로 측정할 것인가를 결정하는 단계
    - 변수의 척도를 고려하여 측정 도구를 설계
    - 객체 : 연구 대상의 단위로 여러 개의 속성으로 구성된다. 예를 들면 자동차 객체는 엔진, 바퀴, 문짝 등이 속성에 해당할 수 있다.
    - 데이터베이스에서 객체는 레코드에 해당하고, 속성은 칼럼에 해당한다. 따라서 칼럼은 변수로 표현되고, 이러한 변수들이 모여서 객체가 된다. 
    - 속성 : 변수 또는 변인. 
    - 변수의 유형 : 독립변수, 종속변수, 매개변수, 조절변수, 외생변수, 억제변수, 왜곡변수
    - 척도 : 연구 대상을 측정하기 위한 측정 도구로서 응답자가 변인의 값을 선택할 수 있도록 일련의 기호 또는 숫자로 나타내어 변수를 측정하게 하는 단위이다. 따라서 척도를 측정 수준이라고도 한다.
    - 명목척도 : 단순 속성 분류, 연산이 의미 없다. 
    - 서열척도 : 순서 관계를 밝혀준다. 연산이 의미 없다. 
    - 등간척도 : 절대원점을 갖고 있지 않으며 측정 대상의 속성에 대한 각 수준 간의 간격이 동일한 척도로 간격이 일정하여 덧셈과 뺄셈 연산이 가능한 변수이다. 
    - 비율척도 : 척도의 수가 등간이며, 절대원점을 가지고 있는 척도를 말한다. 즉, 속성이 0을 기준으로 한 수치로 되어있기 때문에 사칙연산이 모두 가능한 변수이다.
  - 4) 데이터 수집(설문지, 웹, SNS)
    - 선정되 측정 도구를 이용하여 설문 문항을 작성하고, 오프라인 온라인을 통해서 데이터를 수집하는 단계이다.
  - 5) 데이터 코딩/프로그래밍
    - 데이터의 수집을 통해서 획득한 데이터를 통계분석 프로그램(R, Python, SAS, SPSS, Excel 등)을 이용하여 데이터를 입력(코딩)하는 단계이다. 코딩은 사용자의 응답 결과를 숫자나 기호 등을 이용하여 데이터를 입력하는 과정을 말한다.
  - 6) 통계분석 수행(R/SPSS/SAS)
    - 전문 통계분석 프로그램(R,SPSS,SAS)을 이용하여 분석을 시행하는 단계이다. 
    - 변수의 척도에 의해서 대부분 통계분석 방법이 결정되기 때문에 변수의 척도 선정과 모델링이 무엇보다도 중요하다.
    - 빈도분석, 교차분석(카이제곱), 요인 분석, 신뢰도 분석, 상관관계 분석, 회귀분석, t-검정, 분산분석
  - 7) 결과분석(논문/보고서 작성)
    - 연구목적과 연구가설에 대한 검증을 중심으로 분석하는 단계이다. 단순한 통계 결과 제시보다는 인구 통계학적 특성을 시작으로 결과의 의미를 해석하고, 연구자의 개인적인 의견을 기술하여 분석 결과를 제시한다.

- 가설 설정 이전의 연구조사
  - 추론통계 기반의 데이터 분석 절차 이전에 연구조사 과정이 선행되어야 한다. 
  - 연구문제 선정 단계, 예비조사 단계, 연구모형 설계 단계 등이 포함될 수 있다.

- 통계 관련 용어
  - 이용 분야에 따른 통계학의 구분
    - 기술 통계학 : 수집된 자료의 특성을 쉽게 파악하기 위해서 자료를 정리 및 요약
    - 추론 통계학 : 모집단에서 추출한 표본의 정보를 이용하여 모집단의 다양한 특성을 과학적으로 추론

- 전수조사와 표본조사
  - 전수조사 : 모집단 내에 있는 모든 대상을 조사하는 방법
  - 표본조사 : 모집단으로부터 추출된 표본을 대상으로 분석을 시행하기 떄문에 전수조사의 단점을 보완할 수 있지만 모집단의 특성을 반영하는 표본이 추출되지 못하면 수집된 자료가 무용지물이 될 수 있다. 
- 모집단과 표본
  - 모집단은 통계적 관찰 대상이 되는 개체의 전체집합을 의미.
  - 표본은 모집단에서 조사대상으로 추출된 부분집합을 뜻한다. 
  - 모집단으로부터 표본을 추출하는 과정을 표본추출이라고 한다. 
  - 모집단의 특성을 나타내는 모수
  - 표본의 특성을 나타내는 통계량

- 통계적 추정
  - 모집단의 특성을 파악하기 위해서 모집단의 특성을 대표하는 표본을 추출하고, 이러한 표본을 이용하여 모집단의 특성을 나타내는 각종 모수인 모평균, 모분산 등을 예측하는 방법이 통계적 추첮이다. 
    - 점추정 : 모집단의 특성을 하나의 값으로 추정하는 방식, 모수와 동일할 가능성이 가장 큰 하나의 값을 선택하는 방법으로 가능성 희박
    - 구간 추정 : 모집단의 특성을 적절한 구간을 이용하여 추정하는 방식, 모수가 속하는 일정 구간(하한값, 상한값)으로 추정하기 때문에 일반적으로 많이 사용한다. 

- 통계적 추정 中 구간 추정의 주요 용어
  - 신뢰수준 : 계산된 구간이 모수를 포함할 확률
  - 신뢰구간 : 신뢰수준 하에서 모수를 포함하는 구간으로 (하한값, 상한값)의 형식으로 표현한다. 
  - 표본오차 : 모집단에서 추출한 표본이 모집단의 특성과 정확이 일치하지 않아서 발생하는 확률의 차이를 의미한다.

- 기각역과 채택역
  - 이때 귀무가설을 채택하거나 기각하는 기준을 임계값이라고 한다. 
  - 이러한 임계값을 기준으로 채택되는 범위를 채택역이라고 하며, 기각되는 범위를 기각역이라고 표현한다.
  - 즉, 검정 통계량의 값이 기각역에 속하면 귀무가설을 기각하게 된다는 의미이다. 

- 양측 검정과 단측 검정
  - 검정 통계량의 분포에서 유의수준 $\alpha$에 의하여 기각역의 크기가 결정되는데 기각역의 위치는 연구가설의 형태에 의하여 결정된다. 연구가설의 형태에 따라서 양측 검정과 단측 검정으로 나누어진다. 
  - 양측 검정 : 귀무가설이 '성별에 따른 만족도에 차이가 없다'와 같은 형식으로 방향성(어느 한쪽이 많고 적음이 없음)이 없는 경우 적용하는 검정방법이다. 따라서 연구가설은 '성별에 따라 만족도에 차이가 있다'와 같은 방향성이 없는 가설과 방향성(남성의 만족도가 더 크다)을 갖는 형식으로 연구가설을 진술할 수 있다. 
  - 단측 검정 : 어느 한쪽이 많거나 적은 가설인 경우에 해당된다. 
  - 모집단의 특성을 나타내는 수치인 모수가 가설에 의해서 정해지는 특정값보다 작은 경우는 좌측 검정이라고 하고, 모수가 특정값보다 큰 경우에는 우측 검정이라고 한다. 

- 가설 검정 오류
  - 가설을 검정하기 위해서는 귀무가설과 연구가설을 설정하여 통계분석을 통해서 둘 중 하나를 채택하게 된다. 이러한 검정 과정에서 발생할 수 있는 오류는 제 1종 오류와 제 2종 오류가 있다. 
  - 제 1종 오류 : 귀무가설이 참인데도 불구하고 귀무가설을 기각하는 오류를 의미한다. 
  - 제 2종 오류 : 귀무가설이 거짓인데도 불구하고 귀무가설을 채택하는 오류를 의미한다.
  - 제 1종 오류를 범할 확률은 $\alpha$로 표현하고, 제 2종 오류를 범할 확률은 $\beta$로 표현한다. 또한 제 1종 오류를 범할 확률 $\alpha$는 유의수준이라고 하며 제 2종 오류를 범하지 않을 확률은 1-$\beta$로서 검정력이라고 한다. 
  - 검정력이란 대립가설이 맞을 때 귀무가설을 기각할 확률을 의미한다. 

- 가설검정 오류와 기각역은 밀접한 관련이 있다. 
  - 만약 귀무가설의 채택역을 크게 하면 제 1종 오류는 적어지지만 제2종 오류는 커지고, 귀무가설의 채택역을 작게하면 제2종 오류는 작아지지만 제1종 오류가 커지는 특성을 보인다. 
  - 가설검정에서는 이러한 두 가지 오류는 발생할 수밖에 없지만 가능하면 모두 작은 경우가 바람직할 것이다. 따라서 가설검정의 특성에 맞게 적절한 유의수준을 결정하여 가설검정의 오류가 최소화되도록 해야한다.

- 검정 통계량
  - 가설을 검정하기 위해 수집된 데이터로부터 계산된 통계량을 말한다. 즉 가설검정에서 기각역을 결정하는 기준이 되는 통계량을 의미한다. 
  - 이러한 검정 통계량은 유의수준 $\alpha$의 값과 비교하여 귀무가설 기각 혹은 채택 여부를 결정하게 된다. 
  - 검정 통계량은 분석 방법에 따라서 달라지는데 상관분석은 r값, 회귀분석은 F값과 t값, T-검정은 t값, 분산분석은 F값, 카이제곱은 $x^2$값 등으로 나타난다. 

- 예) 연구가설은 '학력 수준에 따라 제품만족도에 차이가 있다'를 검정하기 위해서 독립 표본 T-검정을 수행했고, 이때 유의수준은 0.05로 결정하였다. 검정 결과 검정 통계량 t값이 10.652 유의확률 P값이 0.012가 나왔다고 가정한다면 유의수준 0.05에서 귀무가설이 기각된다. 따라서 학력수준에 따라 제품만족도에 유의미한 차이가 있음을 알 수 있다.
